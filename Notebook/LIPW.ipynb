```python
import pandas as pd
import unicodedata, re

# --- 1 Create a cleaner function to standardize the keys ---
def normalize_key(x):
    """
    Cleans and standardizes class name keys before merging.
    Removes spaces, punctuation, and case differences.
    """
    if pd.isna(x):
        return None
    s = str(x).strip()                    # remove leading/trailing spaces
    s = unicodedata.normalize("NFKD", s)  # normalize unicode (  e)
    s = re.sub(r"\s+", " ", s)            # collapse multiple spaces
    s = s.lower()                         # lowercase for consistency
    s = re.sub(r"[^\w\s]", "", s)         # remove punctuation
    return s

# --- 2 Create cleaned versions of the join columns ---
SubProjectsListing['subProject_clean'] = SubProjectsListing['subProject'].apply(normalize_key)
LIPW_DEVICES['subProject_clean'] = LIPW_DEVICES['subProject'].apply(normalize_key)

# --- 3 Merge both DataFrames ---
merged_df = pd.merge(
    SubProjectsListing,       # left DataFrame
    LIPW_DEVICES,          # right DataFrame
    how='left',               # keep all rows from left side
    on='subProject_clean',     # merge on the cleaned key
    suffixes=('', '_LIPW'),# add suffix to avoid duplicate column names
    indicator=True            # adds a column showing match status
)

# --- 4 Check merge results ---
print(" Merge complete!")
print(merged_df['_merge'].value_counts())  # shows how many matched vs unmatched

# --- 5 Save the result (optional) ---
merged_df.to_csv("merged_BzcosubProject_PICLASS.csv", index=False)
print(" Saved as merged_BzcosubProject_PICLASS.csv")

# --- 6 (Optional) Show sample unmatched rows for inspection ---
unmatched = merged_df[merged_df['_merge'] == 'left_only']['subProject'].unique()[:20]
print("\n Sample unmatched subProject names:", unmatched)
```
