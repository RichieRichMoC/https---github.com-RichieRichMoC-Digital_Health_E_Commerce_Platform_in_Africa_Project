```python
import pandas as pd
import unicodedata, re

# --- 1  Create a safe normalizer function ---
def normalize_key(x):
    """
    Cleans and standardizes keys before merging.
    - Converts to string
    - Removes extra spaces, punctuation
    - Lowercases
    - Normalizes unicode accents
    """
    if pd.isna(x):
        return None
    s = str(x).strip()                   # remove leading/trailing spaces
    s = unicodedata.normalize("NFKD", s) # normalize unicode (e.g.   e)
    s = re.sub(r"\s+", " ", s)           # collapse multiple spaces
    s = s.lower()                        # lowercase everything
    s = re.sub(r"[^\w\s]", "", s)        # remove punctuation/symbols
    return s

# --- 2  Apply normalization to both key columns ---
systemdataset['Beneficiary_clean'] = systemdataset['Beneficiary'].apply(normalize_key)
LEPIP_Beneficiaries_in_Cross_Sector_Dataset['Name_clean'] = LEPIP_Beneficiaries_in_Cross_Sector_Dataset['Name (Full Name)_x'].apply(normalize_key)

# --- 3  Merge using the clean columns ---
merged_df = pd.merge(
    systemdataset,
    LEPIP_Beneficiaries_in_Cross_Sector_Dataset,
    how='left',
    left_on='Beneficiary_clean',
    right_on='Name_clean',
    indicator=True  # shows whether a match was found
)

# --- 4  Inspect match stats ---
print(" Merge done!")
print(merged_df['_merge'].value_counts())

# --- 5  (Optional) Show a few unmatched rows for review ---
unmatched = merged_df[merged_df['_merge'] == 'left_only']['Beneficiary'].unique()[:20]
print("\nSample unmatched beneficiaries:", unmatched)

# --- 6  Save result ---
merged_df.to_csv("merged_dataset_cleaned.csv", index=False)
print(" Saved cleaned merged file: merged_dataset_cleaned.csv")
```
