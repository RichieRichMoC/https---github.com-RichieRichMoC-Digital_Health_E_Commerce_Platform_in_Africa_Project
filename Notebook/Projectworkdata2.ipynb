# Insight 

AGE (Client Age):
The average age of clients is 40.02 years, with a standard deviation of 10.42, indicating a diverse age distribution.

The age ranges from 17 to 98, with most clients between 32 and 47 years old, suggesting the bank primarily engages with working-age individuals, but also includes younger and senior segments.

DURATION (Call Duration in Seconds):
The mean duration of the last contact is 258.32 seconds (~4.3 minutes), with a large standard deviation of 259.31, suggesting high variability in call engagement.

The duration ranges from 0 to 4918 seconds (~82 minutes), showing that while some calls were extremely brief, others were extensive, likely indicating client interest.

 Note: Duration is highly correlated with the outcome (y), but it is only known after the call. Avoid using it in models predicting y before contact.

CAMPAIGN (Contacts in Current Campaign):
Clients were contacted on average 2.57 times, with a standard deviation of 2.77, indicating varied campaign persistence.

Contacts range from 1 to 56, but 75% of clients were contacted 3 times or fewer, suggesting that high-frequency contact is rare.

PDAYS (Days Since Previous Contact):
The average pdays is 962.46, with a very low standard deviation (186.94) because most values are 999  a code indicating the client was not contacted before.

Since 75% of the values are 999, prior contact is uncommon, and this field is more useful as a binary flag (contacted previously or not).

PREVIOUS (Previous Campaign Contacts):
The mean number of past contacts is 0.17, with a maximum of 7, and 75% of clients had 0 previous contacts.

This suggests most clients were new or inactive in previous campaigns.

EMP.VAR.RATE (Employment Variation Rate):
The average employment variation rate is 0.08, with values ranging from -3.4 to 1.4, indicating varying economic conditions during campaigns.

Negative values suggest declining employment environments  possibly affecting client decisions.

CONS.PRICE.IDX (Consumer Price Index):
The average consumer price index is 93.58, with a narrow standard deviation (0.58), suggesting stable prices over the campaign period.

Values range from 92.20 to 94.77, and subtle differences might reflect economic phases that could affect customer sentiment.

CONS.CONF.IDX (Consumer Confidence Index):
The mean confidence index is -40.50, with a standard deviation of 4.63, indicating generally pessimistic consumer sentiment.

Index values vary from -50.8 to -26.9, with more negative values reflecting reduced trust in the economy, possibly influencing deposit decisions.

EURIBOR3M (3-Month Euribor Rate):
The average Euribor rate is 3.62, with a range from 0.63 to 5.05. A wide spread implies changing financial environments during the campaign.

Higher rates could signal tighter credit conditions, which might affect client behavior toward savings or term deposits.

NR.EMPLOYED (Number of Employees):
The average number of employees is 5167.03, with values ranging from 4963.6 to 5228.1.

This feature reflects broader economic conditions (e.g., job creation trends), and its near-steady value range suggests minor fluctuations in employment during the period.

 Summary
These insights reveal the economic context, client demographics, and engagement behavior during the marketing campaigns. Variability in features like duration, campaign contact counts, and economic indicators could have a strong influence on whether a client subscribes to a term deposit.
